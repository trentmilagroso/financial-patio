```python
# Import the learning materials
import yfinance as yf # Dataset
import matplotlib.pyplot as plt # What allows us to graph
import pandas as pd # Allows us to manipulate data
```

```python
# Let's load our AAPL data so we can see what we're working with
apple_data = yf.download('AAPL', start = '2024-06-01', end = '2024-08-09')
print (apple_data)
```

```python
# Our line graph to show us changes in value over time for the AAPL stock, specifically for closing price
plt.figure(figsize=(10, 5))
plt.plot(apple_data['Close'], label='AAPL Close Price', color = 'red')
plt.title('Apple (AAPL) Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Now let's say we want to see the opening price after viewing the closing. 
plt.figure(figsize=(10, 5))
plt.plot(apple_data['Open'], label='AAPL Opening Price', color = 'green')

# Add title and labels
plt.title('Apple (AAPL) Stock Opening Price')
plt.xlabel('Date')
plt.ylabel('Open (USD)')

# Legend, grid, now show plot
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the opening price with a lower z-order (So it's below the red)
plt.figure(figsize=(10, 5))
plt.plot(apple_data['Open'], label='AAPL Opening Price', color='green', zorder=1)

# Plotting the closing price with a higher z-order (So it's above the green)
plt.plot(apple_data['Close'], label='AAPL Close Price', color='red', zorder=2)

# Adding title and labels
plt.title('Apple (AAPL) Stock Opening and Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')

# Adding legend. grid, and the plot
plt.legend()
plt.grid(True)
plt.show()
```

```python


# Assuming data is already defined with 'Close' and 'Open' columns
# Calculate a 30-day moving average
apple_data['30_MA'] = apple_data['Close'].rolling(window=30).mean()

plt.figure(figsize=(10, 5))

# Plotting the opening price
plt.plot(apple_data['Open'], label='AAPL Opening Price', color='green', zorder=1)

# Plotting the closing price
plt.plot(apple_data['Close'], label='AAPL Close Price', color='red', zorder=2)

# Plotting the 30-day moving average
plt.plot(apple_data['30_MA'], label='30-Day Moving Average', linestyle='--', color='blue', zorder=3)

# Adding title and labels
plt.title('Apple (AAPL) Stock Prices with 30-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')

# Adding legend
plt.legend()

# Adding grid
plt.grid(True)

# Display the plot
plt.show()
```
